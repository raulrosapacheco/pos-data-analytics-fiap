## Aquisição e Leitura de Dados

```python
import pandas as pd
```

```python
dados = pd.read_csv("A150850189_28_143_208.csv", encoding="ISO-8859-1",
            skiprows=3, sep=";", skipfooter=12, thousands=".", decimal=",", engine='python')
```


Neste exemplo, o código Python está lendo um arquivo CSV usando o pandas. Aqui estão as opções utilizadas na função `pd.read_csv()`:

- **Caminho do arquivo**: `"A150850189_28_143_208.csv"`: Especifica o caminho do arquivo CSV que está sendo lido.

- **Codificação**: `encoding="ISO-8859-1"`: Usa a codificação ISO-8859-1 (Latin-1) para lidar com caracteres especiais no arquivo CSV.

- **Pular linhas**: `skiprows=3`: Pula as três primeiras linhas do arquivo CSV durante a leitura.

- **Separador de campo**: `sep=";"`: Especifica que o separador de campo no arquivo CSV é `;`.

- **Ignorar linhas finais**: `skipfooter=12`: Ignora as últimas 12 linhas do arquivo CSV durante a leitura.

- **Separadores de milhares e decimais**: `thousands="."` e `decimal=","`: Indicam que o ponto (`.`) é usado como separador de milhares e a vírgula (`,`) é usada como separador decimal nos números do arquivo CSV.

```python
# O método tail() em pandas é usado para visualizar as últimas cinco linhas de um DataFrame ou de uma Série. 
dados.tail()
```

```python
# O método mean() em um DataFrame ou em uma Série do pandas, ele calcula a média dos valores ao longo do eixo especificado. 
dados.mean(numeric_only=True)
```

A configuração `pd.options.display.float_format = "{:.2f}".format` em pandas é usada para definir o formato de exibição dos números de ponto flutuante em um DataFrame.

Quando você aplica essa configuração, todos os números de ponto flutuante no DataFrame serão formatados para terem duas casas decimais após o ponto decimal. É útil para controlar a precisão na exibição dos números, especialmente ao analisar grandes conjuntos de dados.

A instrução `dados.plot(x="Unidade da Federação", y="2008/Ago")` em pandas é usada para criar um gráfico. O eixo x representa os valores da coluna "Unidade da Federação", enquanto o eixo y representa os valores da coluna "2008/Ago" do DataFrame `dados`.

Isso gera um gráfico básico, mostrando a relação entre essas duas colunas. Você pode personalizar o gráfico adicionando títulos, rótulos dos eixos e outros detalhes conforme necessário.


```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

axis = dados.plot(x="Unidade da Federação", y="2008/Ago", kind="bar", figsize=(9,6))
axis.yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.2f}"))

plt.title("Valor por unidade da federação")

plt.show()
```

1. **Importação de Bibliotecas**:
   - `import matplotlib.pyplot as plt`: Importa a biblioteca `matplotlib.pyplot` para criar gráficos.
   - `import matplotlib.ticker as ticker`: Importa o módulo `ticker` de `matplotlib` para formatação de rótulos nos eixos.

2. **Criação do Gráfico de Barras**:
   - `axis = dados.plot(x="Unidade da Federação", y="2008/Ago", kind="bar", figsize=(9,6))`: Cria um gráfico de barras usando os dados da coluna "Unidade da Federação" como eixo x e os valores de "2008/Ago" como eixo y do DataFrame `dados`. O gráfico é do tipo barra e tem o tamanho da figura definido como (9,6).

3. **Formatação dos Rótulos do Eixo Y**:
   - `axis.yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.2f}"))`: Formata os rótulos do eixo y para mostrar números com duas casas decimais e separadores de milhares.

4. **Adição de Título**:
   - `plt.title("Valor por unidade da federação")`: Adiciona o título "Valor por unidade da federação" ao gráfico.

5. **Exibição do Gráfico**:
   - `plt.show()`: Exibe o gráfico na saída.
